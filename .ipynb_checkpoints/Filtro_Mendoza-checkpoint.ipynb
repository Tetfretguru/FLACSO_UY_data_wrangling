```python
respuestas = []
i = 0

while i <= 15:
    respuestas.append(list(diagnostico.iloc[i]))
    i += 1

respuestas
```

```python
preguntas = list(diagnostico)
preguntas
```

```python
# Crear objeto de alumno

class Alumno:
    def __init__(self, aid, preguntas, respuestas):
        self.aid = aid
        self.preguntas = preguntas
        self.respuestas = respuestas
```

```python
Alumno_1 = Alumno(1, preguntas, respuestas[0])

print(Alumno_1.respuestas)
```

**Crear un diccionario con cada alumno para poder visualizar sus respuestas**

```python
i = 0
dir_alumnos = {}

while i <= 15:
    alumno = Alumno(1, preguntas, respuestas[i])
    dir_alumnos[i] = alumno
    i += 1

for k,v in dir_alumnos.items():
    print(f'**Alumno id {k+1}** | Profesor: {v.respuestas[7]}')
    print(v.respuestas)
    print(' ')
    print('-----'*5)
    print(' ')
    
#pd.unique(diagnostico_df['¿Qué temas te gustaría trabajar en el curso?'])
```
